```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
import requests
from pandas_datareader import data as web # Evita a escrita do data e troca pelo web
from datetime import datetime, date
import datetime

print('Hello world')
```

```python
bd="../carteira_v03.sqlite3"
# Consulta ao Banco de dados
def consulta_carteira(tipo):
   
    con=sqlite3.connect(bd)
    df=pd.read_sql_query("select * from carteira",con)

    if tipo == 'Global':
        resumo=pd.read_sql_query("SELECT tipo,sum(Vinvestido) as Distribuição_Carteira from carteira GROUP by tipo",con)
        moeda_carteira=pd.read_sql_query("SELECT moeda,sum(Vinvestido) as total  from carteira GROUP by moeda",con)
        print(df.head())
        return df, resumo, moeda_carteira
    elif tipo == 'Ações':
        ações=pd.read_sql_query("SELECT  ticker,Vinvestido as total  from carteira WHERE tipo = 'Ação'",con)
        ações_setores=pd.read_sql_query("SELECT classe,sum(Vinvestido) as total  from carteira WHERE tipo = 'Ação' GROUP by classe",con)
        ações_moeda=pd.read_sql_query("SELECT moeda,sum(Vinvestido) as total  from carteira WHERE tipo = 'Ação' GROUP by moeda",con)
        ação=pd.read_sql_query("SELECT ticker,quantidade,pm as total  from carteira WHERE subtipo = 'Ação'",con)
        return ações,ações_setores,ações_moeda,ação
    elif tipo == 'FIIs':
        fiis=pd.read_sql_query("SELECT  ticker,Vinvestido as total  from carteira WHERE tipo = 'FII'",con)
        fiis_setores=pd.read_sql_query("SELECT classe,sum(Vinvestido) as total  from carteira WHERE tipo = 'FII' GROUP by classe",con)
        return fiis, fiis_setores
    elif tipo == 'Cripto':
        cript=pd.read_sql_query("SELECT  ticker,Vinvestido as total  from carteira WHERE tipo = 'Cripto'",con)
        return cript
    elif tipo == 'Tesouro':
        td=pd.read_sql_query("SELECT  ticker,classe,Vinvestido as total  from carteira WHERE tipo = 'TD'",con)
        td_setores=pd.read_sql_query("SELECT classe,sum(Vinvestido) as total  from carteira WHERE tipo = 'TD' GROUP by classe",con)
        return td, td_setores
   
    con.close()
    
#resumo=consulta_carteira()
```

```python
def atualiza_FIIs():
    # Primeiramente, obtém-se dados atualizados dos fundos de investimento
    # Alguns ajustes são feitos na tabela para a criação dos dataframes apropriados
    url = 'https://www.fundsexplorer.com.br/ranking'
    headers = {
    'User-Agent': 
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_4) AppleWebKit/537.36'
        ' (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        dfii = pd.read_html(response.content, encoding='utf-8')[0]
    dfii.sort_values('Códigodo fundo', inplace=True)

    dfii
    col_floats = list(dfii.iloc[:,2:-1].columns)
    dfii[col_floats] = dfii[col_floats].fillna(value=0)
    dfii[col_floats] = dfii[col_floats].applymap(lambda x: str(x).replace('R$', '').replace('.0','').replace('.','').replace('%','').replace(',','.'))
    dfii[col_floats] = dfii[col_floats].astype('float')

    # Open connection to BD
    fiis, fiis_setores=consulta_carteira('FIIs')
    lista=np.array(fiis.ticker)
    con=sqlite3.connect("../carteira_v03.sqlite3")
    cursor = con.cursor()

#    print(fiis.head())
    sql="""UPDATE carteira SET Patual = ? WHERE id = ?"""
    for ticker in lista:
        query="select id from carteira WHERE ticker = '"+str(ticker)+"' "
        ticker_id=np.array(pd.read_sql_query(query,con))
        fii_price=float(np.array(dfii.loc[dfii['Códigodo fundo']==str(ticker)])[0,2])#/100.0
        var=[str(fii_price),str(int(ticker_id))]
        print(ticker,str(fii_price))
        cursor.execute(sql,var)
        con.commit()
    #df=pd.read_sql_query("select ticker, Patual from carteira",con)
    con.close()
    print('FIIs atualizados ...')


#df.head()
```
